```python
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import numpy as np
import torch
from tqdm import tqdm
import os
from PIL import Image
from sklearn.decomposition import PCA
import torchvision.transforms.functional as F


def adapt_dataset(origin_directory_path, destination_directory_path, property_value, adaptation_method):
    # Walk through the source directory
    paths = []
    destination_paths = []

    for root, _, files in os.walk(origin_directory_path):
        for file in files:
            # Construct full file path
            file_path = os.path.join(root, file)
            paths.append(file_path)

            destination_path = os.path.join(destination_directory_path, os.path.relpath(file_path, origin_directory_path))
            destination_paths.append(destination_path)

    paths.sort()
    destination_paths.sort()

    # Initialize the progress bar
    progress_bar = tqdm(total=len(paths), desc=f'Processing files in {origin_directory_path}')
    print("Property value: ", property_value)

    for path, destination_path in zip(paths, destination_paths):
        adaptation_method(path, destination_path, property_value)
        progress_bar.update(1)

    # Close the progress bar
    progress_bar.close()

def adjust_noise(image_path, destination_path, noise_factor):
    image = Image.open(image_path)
    noisy_image = np.array(image)
    if np.max(noisy_image) > 1 and noise_factor <= 1:
        noise_factor *= 255
    noise = np.random.normal(0, noise_factor, noisy_image.shape)
    noisy_image = np.clip(noisy_image + noise, 0, 255)

    noisy_image = Image.fromarray(noisy_image.astype(np.uint8))

    # Save the image
    noisy_image.save(destination_path)

    return noise_factor

def adjust_brightness(image_path, destination_path, brightness_factor):
    image = Image.open(image_path)
    brightened_image = F.adjust_brightness(image, brightness_factor)

    # Save the image
    brightened_image.save(destination_path)

    return brightness_factor
            
def flatten_features(features):
    """
    Flattens the features for t-SNE.

    Args:
        features (torch.Tensor): Tensor of shape (B, C, H, W) or (B, C).

    Returns:
        np.ndarray: Flattened features of shape (B, C*H*W) or (B, C).
    """
    if len(features.shape) == 4:
        # For features of shape (B, C, H, W), flatten to (B, C*H*W)
        B, C, H, W = features.shape
        flattened = features.view(B, -1)  # Shape: (B, C*H*W)
    elif len(features.shape) == 2:
        # For features already in (B, C) shape, no need to flatten further
        flattened = features
    else:
        raise ValueError("Unsupported feature shape")

    return flattened.cpu().detach().numpy()  # Convert to NumPy for t-SNE

def reduce_features_tsne(features, n_components=2, perplexity=30):
    """
    Apply t-SNE on flattened features.

    Args:
        features (torch.Tensor): Tensor of shape (B, C, H, W) or (B, C).
        n_components (int): Number of dimensions to reduce to.
        perplexity (int): t-SNE perplexity parameter.
        
    Returns:
        np.ndarray: 2D array of reduced features.
    """
    # Flatten the features
    flattened_features = flatten_features(features)
    
    # Apply t-SNE to reduce dimensionality
    perplexity = min(perplexity, flattened_features.shape[0] - 1)  # Ensure perplexity < number of samples
    tsne = TSNE(n_components=n_components, random_state=42, perplexity=perplexity)
    reduced_features = tsne.fit_transform(flattened_features)
    
    return reduced_features

def reduce_features_pca(features, n_components=2):
    """
    Apply PCA to flatten and reduce features.

    Args:
        features (torch.Tensor): Tensor of shape (B, C, H, W) or (B, C).
        n_components (int): Number of dimensions to reduce to.
        
    Returns:
        np.ndarray: 2D array of reduced features.
    """
    flattened_features = flatten_features(features)
    
    # Apply PCA for dimensionality reduction
    pca = PCA(n_components=n_components)
    reduced_features = pca.fit_transform(flattened_features)
    
    return reduced_features

def get_tsne_plots():
    reduced_features_x = []
    reduced_features_x_e = []
    reduced_features_output = []

    model.model.eval()

    accuracy_scores = 0.0

    progress_bar = tqdm(total=len(dataloaders['val']), desc='Processing validation data')

    for i, minibatch in enumerate(dataloaders['val']):
        images = minibatch['data']
        modal_xs = minibatch['modal_x']
        label = minibatch['label']

        if len(images.size()) == 3:
            images = images.unsqueeze(0)
            modal_xs = modal_xs.unsqueeze(0)
            label = label.unsqueeze(0)

        with torch.no_grad():
            # Forward pass through the model's backbone for each modality
            x = model.model.backbone.downsample_layers[0](images)
            x_e = model.model.backbone.downsample_layers_e[0](modal_xs)

            # Apply dimensionality reduction (t-SNE or PCA) after each batch
            # reduced_x = reduce_features_tsne(x)
            # reduced_x_e = reduce_features_tsne(x_e)
            reduced_x = reduce_features_pca(x)
            reduced_x_e = reduce_features_pca(x_e)

            # Get model's output and apply t-SNE or PCA
            output = model.model(images, modal_xs)
            # reduced_output = reduce_features_tsne(output)
            reduced_output = reduce_features_pca(output)

            # Append reduced features to the lists
            reduced_features_x.append(reduced_x)
            reduced_features_x_e.append(reduced_x_e)
            reduced_features_output.append(reduced_output)

            accuracy = (output.softmax(1).argmax(1) == label).float().mean().item()
            accuracy_scores += accuracy
        
        progress_bar.update(1)

    # Concatenate the reduced features for all batches
    reduced_x = np.vstack(reduced_features_x)
    reduced_x_e = np.vstack(reduced_features_x_e)
    reduced_output = np.vstack(reduced_features_output)

    print(f"Average accuracy: {accuracy_scores / len(dataloaders['val'])}")

    return reduced_x, reduced_x_e, reduced_output


origin_directory_path=r"datasets\NYUDepthv2\RGB_original"
destination_directory_path=r"datasets\NYUDepthv2\RGB"


adapt_dataset(
    origin_directory_path=origin_directory_path,
    destination_directory_path=destination_directory_path,
    property_value=0.0,
    adaptation_method=adjust_brightness,
)

reduced_x_noise, reduced_x_e_noise, reduced_output_noise = get_tsne_plots()


adapt_dataset(
    origin_directory_path=origin_directory_path,
    destination_directory_path=destination_directory_path,
    property_value=1.0,
    adaptation_method=adjust_brightness,
)

reduced_x_no_noise, reduced_x_e_no_noise, reduced_output_no_noise = get_tsne_plots()

```

```python
def plot_tsne(reduced_features_a, reduced_features_b, title):
    """
    Plot 2D reduced features (after t-SNE or PCA).
    
    Args:
        reduced_features (np.ndarray): 2D array of shape (B, 2) after t-SNE.
        title (str): Title of the plot.
    """
    # Plot the reduced features (2D after t-SNE or PCA)
    plt.figure(figsize=(8, 6))
    # plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=np.arange(reduced_features.shape[0]), cmap='viridis', label='Samples')
    plt.scatter(reduced_features_a[:, 0], reduced_features_a[:, 1], c='r', label='Noise')
    plt.scatter(reduced_features_b[:, 0], reduced_features_b[:, 1], c='b', label='No Noise')
    plt.colorbar(label='Sample Index')
    plt.title(f'{title}')
    plt.xlabel('Reduced Dim 1')
    plt.ylabel('Reduced Dim 2')
    plt.legend()
    plt.show()

plot_tsne(reduced_x_noise, reduced_x_no_noise, title='Feature Map (RGB) PCA')
plot_tsne(reduced_x_e_noise, reduced_x_e_no_noise, title='Feature Map (Depth) PCA')

plot_tsne(reduced_output_noise, reduced_output_no_noise, title='Output Feature Map PCA')
```
